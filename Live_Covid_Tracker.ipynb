# Corona Virus Live Tracking For India

```python
import pandas as pd

df = pd.read_json('https://covid19dashboard.mohfw.gov.in/data/datanew.json')
df
```

```python
df=df.iloc[:-1,1:]
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
import numpy as np
total_states = np.arange(len(df['state_name']))
total_states
```

```python
df['active']
```

```python
max(df['positive'])
```

# Total Positive Cases Based On States In India

```python
plt.barh(total_states,df['positive'])
plt.xlim(1,max(df['positive'])+100)
plt.yticks(total_states,df['state_name'])
plt.xlabel('Total Positive Cases')
```

# Active Number of Cases based on States

```python
plt.barh(total_states,df['active'])
plt.xlim(1,max(df['active'])+100)
plt.yticks(total_states,df['state_name'])
plt.xlabel('Total Active Cases')
```

# Death Number of Cases based on States

```python
plt.barh(total_states,df['death'])
plt.xlim(1,max(df['death'])+100)
plt.yticks(total_states,df['state_name'])
plt.xlabel('Total Death Cases')
```

# New Active Number of Cases based on States

```python
plt.barh(total_states,df['new_active'])
plt.xlim(1,max(df['new_active'])+100)
plt.yticks(total_states,df['state_name'])
plt.xlabel('Total New Active Cases')
```
